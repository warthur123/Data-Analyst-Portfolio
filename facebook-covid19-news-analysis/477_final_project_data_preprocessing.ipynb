```python
import pandas as pd
import numpy as np
```

```python
#load datasets
cnn = pd.read_csv('/Users/arthurwu/Downloads/datasets/cnn_fb_data.csv')
nyt = pd.read_csv('/Users/arthurwu/Downloads/datasets/nyt_fb_data.csv')
washpost = pd.read_csv('/Users/arthurwu/Downloads/datasets/wash_fb_data.csv')
nbc = pd.read_csv('/Users/arthurwu/Downloads/datasets/nbc_fb_data.csv')
fox = pd.read_csv('/Users/arthurwu/Downloads/datasets/fox_fb_data.csv')
bbc = pd.read_csv('/Users/arthurwu/Downloads/datasets/bbc_fb_data.csv')
wallstreet = pd.read_csv('/Users/arthurwu/Downloads/datasets/wsj_fb_data.csv')
abc = pd.read_csv('/Users/arthurwu/Downloads/datasets/abcnews_fb_data.csv')
guardian = pd.read_csv('/Users/arthurwu/Downloads/datasets/theguardian_fb_data.csv')
usatoday = pd.read_csv('/Users/arthurwu/Downloads/datasets/usatoday_fb_data.csv')

dataset_list = [cnn, nyt, washpost, nbc, fox, bbc, wallstreet, abc, guardian, usatoday]
```

```python
usatoday.info()
```

```python
# convert numeric columns to numeric
for df in dataset_list:
    df[['# Shares', '# Total Reactions', '# Like', '# Love', '# Haha', '# Wow', '# Sad', '# Angry', '# Comments']] = df[['# Shares', '# Total Reactions', '# Like', '# Love', '# Haha', '# Wow', '# Sad', '# Angry', '# Comments']].astype(float)
```

```python
fb_data = pd.concat(dataset_list)
fb_data.info()
```

```python
# drop invalid rows
fb_data = fb_data.dropna(subset=['Posted By', '# Shares', '# Total Reactions'])
```

```python
fb_data.info()
```

```python
# filter for rows that contain covid keywords in post title and post text columns
fb_data = fb_data[fb_data["Post's Title"].str.contains("covid", case=False, regex=False) |
                  fb_data["Post's Title"].str.contains("coronavirus", case=False, regex=False) |
                  fb_data["Post's Title"].str.contains("pandemic", case=False, regex=False) |
                  fb_data["Post's Title"].str.contains("vaccine", case=False, regex=False) |
                  fb_data["Post's Text"].str.contains("covid", case=False, regex=False) |
                  fb_data["Post's Text"].str.contains("coronavirus", case=False, regex=False) |
                  fb_data["Post's Text"].str.contains("pandemic", case=False, regex=False) |
                  fb_data["Post's Title"].str.contains("vaccine", case=False, regex=False)
                 ]
```
